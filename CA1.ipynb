Given that most job postings are from the United States, we can set the number of work hours per day in the US as a standard. I will use this to calculate and standardize the yearly salaries. I've learned that Full-time is the most common work type, so I will use 40 hours as the standard workweek to transform the salary-columns.

```python
def impute_salaries(row):
    # Adjust salary based on 'pay_period'
    if row['pay_period'] == 'HOURLY':
        factor = 40 * 52  # 40 hours/week * 52 weeks/year
    elif row['pay_period'] == 'MONTHLY':
        factor = 12  # 12 months in a year
    elif row['pay_period'] == 'WEEKLY':
        factor = 52  # 52 weeks in a year
    elif row['pay_period'] == 'BIWEEKLY':
        factor = 26  # 26 biweekly periods in a year
    elif row['pay_period'] == 'YEARLY':
        factor = 1  # Already yearly, no adjustment needed
    else:
        factor = 1  # Default to 1 for unexpected pay_period values

    # Adjust 'med_salary', 'max_salary', and 'min_salary' accordingly
    if row['med_salary'] is not None:  # If the value is not None
        row['med_salary'] *= factor
    if row['max_salary'] is not None:  # If the value is not None
        row['max_salary'] *= factor
    if row['min_salary'] is not None:  # If the value is not None
        row['min_salary'] *= factor

    return row
# Apply the function to the entire dataframe
it_df = it_df.apply(impute_salaries, axis=1)
```

As we can observe, there is a positive relationship between 'Views' and 'Applies'. Therefore, I will use this relationship to impute the missing values. I will calculate the ratio of 'Applies' to 'Views' for each data point and then compute the average of this ratio, if the distribution is not too skewed.

```python
# Define a function to calculate the ratio of 'applies' to 'views'
def calculate_ratio(row):
    if pd.notnull(row['applies']) and pd.notnull(row['views']):
        return row['applies'] / row['views']
    else:
        return np.nan

# Apply the function to create the new column
it_df['applies_to_views_ratio'] = it_df.apply(calculate_ratio, axis=1)

```

To impute the missing values in the 'Applies' column, I need to impute the 'Views' column first. I will impute based on 'formatted_work_type'.

```python
# Group the data by 'formatted_work_type' and calculate the median of views for each work type
median_views = it_df.groupby('formatted_work_type')['views'].median().round()

# Store the median values in a variable
print(median_views)

# Now, you can use this 'median_views' to impute missing values for each 'formatted_work_type'
# Create the bar graph to visualize the median views by work type
plt.figure(figsize=(10, 6))
median_views.plot(kind='bar', color='#d62728')

# Adding title and labels
plt.title('Median Views by Work Type')
plt.ylabel('Median Views')

# Set x-ticks for work types
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()

# Store the median values in the DataFrame for further use (for imputing)
it_df['median_views'] = it_df['formatted_work_type'].map(median_views)

```

Based on the result above, I will impute the missing values in the 'Views' column with a scalar value according to formatted_work_type.

```python
# Define the function to impute 'views' based on 'formatted_work_type'
def impute_views(row):
    # Check the 'formatted_work_type' and impute 'views' based on the work type
    if row['formatted_work_type'] == 'Contract':
        if pd.isna(row['views']):
            row['views'] = 10.0
    elif row['formatted_work_type'] == 'Full-time':
        if pd.isna(row['views']):
            row['views'] = 4.0
    elif row['formatted_work_type'] == 'Internship':
        if pd.isna(row['views']):
            row['views'] = 6.0
    elif row['formatted_work_type'] == 'Other':
        if pd.isna(row['views']):
            row['views'] = 4.0
    elif row['formatted_work_type'] == 'Part-time':
        if pd.isna(row['views']):
            row['views'] = 4.0
    elif row['formatted_work_type'] == 'Temporary':
        if pd.isna(row['views']):
            row['views'] = 4.0
    else:
        # Default to 4.0 for any unlisted work type
        if pd.isna(row['views']):
            row['views'] = 4.0
    return row

# Apply the function to the entire dataframe
it_df = it_df.apply(impute_views, axis=1)

# Verify that missing values have been imputed
print(it_df['views'].isnull().sum())
```

Now, it's time to impute the 'applies' column using the 'median_ratio'. 

```python
# Define the function to impute 'applies' using 'median_ratio'
def impute_applies(row, median_ratio):
    if pd.isnull(row['applies']) and pd.notnull(row['views']):
        # Impute the missing 'applies' value and round it
        return round(row['views'] * median_ratio)
    else:
        return row['applies']

# Apply the function to impute 'applies'
it_df['applies'] = it_df.apply(impute_applies, axis=1, median_ratio=median_ratio)

# Check if there are any remaining missing values in 'applies'
print(it_df['applies'].isnull().sum())

```

```python
# Define built-in colors from Matplotlib
custom_colors = ['lightblue', 'skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightpink']

# Filter the dataset for entry-level jobs and count work type distribution
work_type_distribution = it_df[it_df['formatted_experience_level'] == 'Entry level']['formatted_work_type'].value_counts()

# Define a function to remove percentages for small slices
def remove_small_slices(pct, threshold=5):
    return f'{pct:.1f}%' if pct > threshold else ''

# Create the pie chart
plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(work_type_distribution, autopct=lambda pct: remove_small_slices(pct), 
                                   labels=None, colors=custom_colors)

# Add title and legend with percentages
plt.title('Entry-Level Jobs by Work Type', fontsize=14)
plt.legend(wedges, [f'{label} ({count}, {count / work_type_distribution.sum() * 100:.1f}%)' 
                    for label, count in work_type_distribution.items()], 
           title='Work Types', loc='center left', bbox_to_anchor=(1, 0.5))

# Adjust layout and show the plot
plt.tight_layout()
plt.show()

```
